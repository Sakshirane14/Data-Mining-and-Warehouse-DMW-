```python
numeric_cols = ['Age', 'Salary', 'Experience', 'Score']
df1[numeric_cols] = df1[numeric_cols].fillna(df1[numeric_cols].median())
df1.isnull().sum()
```

```python
for col in numeric_cols:
    df1[col + '_zscore'] = (df1[col] - df1[col].mean()) / df1[col].std()

display(df1.head())
```

```python
from sklearn.preprocessing import MinMaxScaler, RobustScaler, StandardScaler

# Initialize scalers
min_max_scaler = MinMaxScaler()
robust_scaler = RobustScaler()
standard_scaler = StandardScaler()


df_minmax = pd.DataFrame(min_max_scaler.fit_transform(df1[numeric_cols]), columns=[col+'_minmax' for col in numeric_cols])
df_robust = pd.DataFrame(robust_scaler.fit_transform(df1[numeric_cols]), columns=[col+'_robust' for col in numeric_cols])
df_standard = pd.DataFrame(standard_scaler.fit_transform(df1[numeric_cols]), columns=[col+'_standard' for col in numeric_cols])

print("\nMin-Max Scaled Data (first 5 rows):\n", df_minmax.head())
print("\nStandard Scaled Data (first 5 rows):\n", df_standard.head())
print("\nRobust Scaled Data (first 5 rows):\n", df_robust.head())
```

```python
# Central Tendency and Dispersion
central_dispersion = {
    'Mean': df1[numeric_cols].mean(),
    'Median': df1[numeric_cols].median(),
    'Mode': df1[numeric_cols].mode().iloc[0],
    'Std Dev': df1[numeric_cols].std(),
    'Variance': df1[numeric_cols].var(),
    'Range': df1[numeric_cols].max() - df1[numeric_cols].min(),
    'IQR': df1[numeric_cols].quantile(0.75) - df1[numeric_cols].quantile(0.25)
}
central_dispersion_df = pd.DataFrame(central_dispersion)
print("\nCentral Tendency and Dispersion:\n", central_dispersion_df)
```
